# Formatting Data – making data types compaitable with other data types

'TOTAL TRADED QUANTITY','TURNOVER' and 'NO. OF TRADE' are the only column wih different data types

```python
#coverting 'TOTAL TRADED QUANTITY' and 'NO. OF TRADES' to float.
data['TOTAL TRADED QUANTITY'] = data['TOTAL TRADED QUANTITY'].astype(float)
data['NO. OF TRADES'] = data['NO. OF TRADES'].astype(float)
```

```python
# Checking if the values changed correctly or not
data
```
