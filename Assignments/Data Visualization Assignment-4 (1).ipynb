## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
data=pd.DataFrame({"Store":["Delhi","Mumbai","Chennai","Kolkata","Hyderabad"],
                  "Clothing":[82,78,48,46,39],
                  "Equipment":[48,31,51,29,18],
                  "Accessories":[15,20,14,13,9]})
```

```python
data.head()
```

```python
var=np.arange(5)
width=0.2
plt.bar(var-width,data['Clothing'],width,label="Clothing")
plt.bar(var,data['Equipment'],width,label="Equipment")
plt.bar(var+width,data['Accessories'],width,label="Accessories")
plt.legend()
plt.xlabel("Store")
plt.xticks(var,list(data["Store"]),rotation=90)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
tt=titanic.groupby(["pclass","survived"]).count()
not_survived=[tt.iloc[i,:].max() for i in [0,2,4]]
survived=[tt.iloc[i,:].max() for i in [1,3,5]]
```

```python
width=0.2
var=np.arange(1,4)
plt.bar(var-width/2,survived,width,label="Survived")
plt.bar(var+width/2,not_survived,width,label="Not Survived")
plt.xticks(var,["pclass-1","pclass-2","pclass-3"])
plt.xlabel("class")
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
sns.barplot(x='class',y='survived',hue='class',data=titanic,palette="spring",)
plt.legend(loc="upper left")
plt.show()
```
